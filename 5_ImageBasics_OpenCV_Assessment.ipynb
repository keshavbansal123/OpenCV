#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
from PIL import Image

img = cv2.imread('dog_backpack.png')
type(img)
```

```python
plt.imshow(img)
```

```python
# image from matplotlib is not in proper color format so needs to change from BGR (OpenCV) to RGB

fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
flip_img = cv2.flip(fix_img,0)
plt.imshow(flip_img)
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
cv2.rectangle(fix_img,pt1=(200,380),pt2=(600,710),
              color=(255,0,0),thickness=10)
plt.imshow(fix_img)
```
